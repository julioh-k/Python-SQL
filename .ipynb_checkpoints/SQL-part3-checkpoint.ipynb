````python
#Libraries 
import sqlite3
import pandas as pd

#connection with db
conn = sqlite3.connect("ladder.db")
````

````python
# 62) Look at the `yum` table. It is the stock data for Yum! Brands, Inc. from 2015 through 2019. Yum! is the 
# company that owns Taco Bell, the best restaurant.

sql = '''SELECT * FROM Yum'''
pd.read_sql(sql,conn)
````

````python
# 63) Query the `yum` table, aggregating by **both** month and year, with the following resulting columns:
# * Year (4 digits)
# * Month
# * Average open, high, low, and close
# * Total volume
# Finally, sort this data so it's in proper chronological order.

sql = '''SELECT SUBSTR(date,1,4) AS Year ,SUBSTR(date,6,2) AS Month , AVG(open) AS Average_open, AVG(high) AS Average_high, AVG(low) AS Average_low, AVG(close) AS Average_close, SUM(volume) AS Total_volume
        FROM Yum
        GROUP BY Year, Month;'''
pd.read_sql(sql,conn)
````

````python
# 64) Save the results of the previous query as a view named `yum_by_month`.

sql = '''CREATE VIEW yum_by_month AS
        SELECT SUBSTR(date,1,4) AS Year ,SUBSTR(date,6,2) AS Month , AVG(open) AS Average_open, AVG(high) AS Average_high, AVG(low) AS Average_low, AVG(close) AS Average_close, SUM(volume) AS Total_volume
        FROM Yum
        GROUP BY Year, Month;'''
pd.read_sql(sql,conn)
````

````python
sql = '''SELECT * FROM yum_by_month'''
pd.read_sql(sql,conn)
````

````python
# 65) Create a view of `transactions` consisting of only three columns: year, month, and total sales in that month.
# Call this view `trans_by_month`.

sql = '''CREATE VIEW trans_by_month AS 
        SELECT SUBSTR(orderdate,1,4) AS Year, SUBSTR(orderdate,6,2) AS Month, SUM(unit_price*quantity) AS Sales
        FROM transactions
        GROUP BY Year, Month;'''
pd.read_sql(sql,conn)
````

````python
sql = '''SELECT * FROM trans_by_month'''
pd.read_sql(sql,conn)
````

````python
# 66) Create a view of `transactions` consisting of only two columns: `employee_id` and the total sales corresponding 
# to that employee. Call this view `trans_by_employee`.

sql=''' CREATE VIEW trans_by_employee AS
        SELECT employee_id, SUM(unit_price*quantity)
        FROM transactions
        GROUP BY employee_id;'''
pd.read_sql(sql,conn)
````

````python
sql = '''SELECT * FROM trans_by_employee'''
pd.read_sql(sql,conn)
````

````python
# 67) What's the most common first initial for pets in the `pets` table?
#     * _Hint:_ Create a CTE that is simply the lowercased first letter of the pet's name. The solution is a simple `GROUP BY` from this CTE.
#     * _Hint 2:_ You'll need the `SUBSTR()` and `LOWER()` functions.

sql = '''WITH letter_CTE AS(
        SELECT LOWER(SUBSTR(name,1,1)) AS letter
        FROM pets
        )
        SELECT letter, COUNT(letter) AS number_frecuncy 
        FROM letter_CTE
        GROUP BY letter
        ORDER BY number_frecuncy DESC
        LIMIT 1;'''
pd.read_sql(sql,conn)
````

````python
# 68) Create taglines for each employee in the `employees` table. As a template, the first row of the result should look like this:
# ```
# Christine Thompson started in 2005 and makes $123,696 working in sales. 
# ```

sql = '''WITH employees_CTE AS(
        SELECT *
        FROM employees
        )
        SELECT firstname||' '|| lastname || ' started in '|| SUBSTR(startdate,1,4) ||' and makes '|| 
        printf('$%,.2d',salary) || ' working ' || job AS taglines
        FROM employees_CTE;'''
pd.read_sql(sql,conn)
````
